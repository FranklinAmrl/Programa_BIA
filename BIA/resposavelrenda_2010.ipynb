```python
import numpy as np
```

```python
import pandas as pd
```

```python
responsavelrenda = pd.read_excel("dados_entrada\Tabelas_2010\ResponsavelRenda_PE_M.xlsx")
```

```python
responsavel02 = pd.read_excel("dados_entrada\Tabelas_2010\Responsavel02_PE_M.xlsx",index=False)
```

```python
variaveis02 = pd.DataFrame({'V001_02': responsavel02['V001']})
```

```python
responsavelrenda = pd.concat([responsavelrenda, variaveis02], axis=1)
```

```python
def V21(V076,V001):
    return ((V076/V001)*100)
```

```python
def V22(V067,V068,V069,V070,V001):
    return (((V067+V068+V069+V070)/V001)*100)
```

```python
resultado = pd.read_excel("dados_saida\Resultado_2010\excel\ResultadoGeral.xlsx",index=False)
```

```python
resultado['V21'] = responsavelrenda.apply(lambda x: V21(x.V076,x.V001_02), axis=1)
```

```python
resultado['V22'] = responsavelrenda.apply(lambda x: V22(x.V067,x.V068,x.V069,x.V070,x.V001_02), axis=1)
```

```python
resultado = resultado.sort_index(axis=1)
```

```python
resultado.to_excel("dados_saida\Resultado_2010\excel\ResultadoGeral.xlsx",index=False)
```

```python
resultado.to_csv("dados_saida\Resultado_2010\csv\ResultadoGeral.csv",index=False)
```
